```python
short_rolling = data['Close Price'].rolling(window=21).mean()
```

```python
long_rolling = data['Close Price'].rolling(window=34).mean()
```

```python
start_date = data.index[0]
end_date = data.index[-1]

fig, ax = plt.subplots(figsize=(16,9))

ax.plot(data.loc[data.index[0]:data.index[-1]].index, data.loc[data.index[0]:data.index[-1]]['Close Price'], label='Price')
ax.plot(data.loc[data.index[0]:data.index[-1]].index, long_rolling.loc[start_date:end_date], label = '34-days SMA')
ax.plot(data.loc[data.index[0]:data.index[-1]].index, short_rolling.loc[start_date:end_date], label = '21-days SMA')

ax.legend(loc='best')
ax.set_ylabel('Price in $')
```

```python
rolling_mean = data['Close Price'].rolling(window=21).mean()
```

```python
rolling_mean.loc[start_date:end_date].std()
```

```python
lower=rolling_mean-2*rolling_mean.loc[start_date:end_date].std()
upper=rolling_mean+2*rolling_mean.loc[start_date:end_date].std()
```

```python
# set style, empty figure and axes
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)

# Get index values for the X axis for facebook DataFrame
x_axis = data.loc[data.index[0]:data.index[-1]].index

# Plot shaded 21 Day Bollinger Band for Facebook
ax.fill_between(x_axis, upper, lower, color='orange')

# Plot Adjust Closing Price and Moving Averages
ax.plot(x_axis, data['Close Price'], color='blue', lw=2,label='Close Price')
ax.plot(x_axis, rolling_mean, color='black', lw=2,label='14 Day Rolling Mean')
ax.legend()
plt.show();
```
